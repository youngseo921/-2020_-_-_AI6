```python
from selenium import webdriver
from selenium.common.exceptions import TimeoutException  # 
from selenium.webdriver.common.keys import Keys
from tqdm.notebook import tqdm #진행상황표시바
import time

driver = webdriver.Chrome(executable_path='/Users/justine/chromedriver')

url = "https://www.google.com"

driver.get(url)
#-----------------------------구글켜기------------------------------------------
google = driver.find_element_by_name('q')# 구글검색창

google.send_keys('ssg')#구글검색창 입력
google.submit()#enter
#-----------------------------SSG------------------------------------------
driver.find_element_by_xpath('//*[@id="tads"]/div/div/div/div[1]/a/div[1]').click()#링크접속

ssg2= driver.find_element_by_class_name('inp_txt')#검색창
ssg2.click()
ssg2.send_keys('중국당면')#입력

ssg3 = driver.find_element_by_id('ssg-query-btn')#enter
ssg3.click()
#----------------------------곱창중 첫번째 -------------------------------------------
time.sleep(4)

driver.find_element_by_xpath('//*[@id="item_unit_1000034736488"]/div[1]/div[2]/a').send_keys(Keys.ENTER)
#이동

#-----------------------------리뷰-------------------------------------------
review = driver.find_elements_by_css_selector('#cdtl_cmt_tbody>tr')#리뷰 파싱

""" 리뷰 1번 페이지 
review = driver.find_elements_by_css_selector('#cdtl_cmt_tbody>tr')

review_data_list=[]

for review_data in tqdm(review):
    print(review_data.text.split('\n'))
    print('--------------------')
    review_data_list.append(review_data.text) 
    
last = driver.find_elements_by_css_selector('#cdtl_cmt_tbody>tr')
print(last[0].text) #리뷰 1번페이지 1번째
""" 
review_data_list=[]
#-----------------------------반복문-----------------------------------------
#반복문 돌리는데 문제점 - 페이지 번호가 xpath 번호랑 다름
#//*[@id="comment_navi_area"]/a[1] xpath로 페이지 번호 1번으로 이동해야하는데 2번으로 움직이고
#2번 페이지번호는 1번페이지로 이동, 3번 페이지번호는 4번페이지, 4번페이지번호는 3번으로 넘어감, 5번부터는 정상적
#이 문제점을 해결하기 위해서 처음부터 9번페이지로 넘어가서 
#8번부터 시작하면 번호는 달라도 어쨋든 모든페이지번호에있는 리뷰 가져오기 가능

time.sleep(4) #no such element: Unable to locate element 라는 오류가 아직 페이지를 다 불러오지 못했기때문에 기다려야함
driver.find_element_by_xpath('//*[@id="comment_navi_area"]/a[8]').send_keys(Keys.ENTER)
#8번 페이지번호로 넘어가서 시작 

#——————————————페이지1번부터10번까지————————————————————
for i in tqdm(range(1,10+1)):
    f = '//*[@id="comment_navi_area"]/a[{}]'.format(str(i))
    time.sleep(1)
    driver.find_element_by_xpath(f).send_keys(Keys.ENTER)
    time.sleep(3)
    
    rev = driver.find_elements_by_css_selector('#cdtl_cmt_tbody>tr') #처음 정의했던걸 다시 정의하고 시작하니 되는거같습니다!!
    
    for review_data in tqdm(rev):
        review_data_list.append(review_data.text.split('\n'))
        print('—————————2')
        time.sleep(3)
        
#————————————————페이지 n번에 있는 리뷰 10개 불러오기——————————————//*
for h in range(1,999):
    for i in tqdm(range(3,12+1)):
        time.sleep(1)
        f = '//*[@id="comment_navi_area"]/a[{}]'.format(str(i))
        print(driver.find_element_by_xpath(f).text)
        driver.find_element_by_xpath(f).send_keys(Keys.ENTER)
        time.sleep(3)

        rev = driver.find_elements_by_css_selector('#cdtl_cmt_tbody>tr') #처음 정의했던걸 다시 정의하고 시작하니 되는거같습니다!!

        for review_data in tqdm(rev):
            print(review_data.text.split('\n'))
            review_data_list.append(review_data.text.split('\n'))
            print('—————————2')
            time.sleep(3)
        
        
        
```
