```python
from selenium import webdriver
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.common.keys import Keys
from tqdm.notebook import tqdm #진행상황표시바
import time

driver = webdriver.Chrome(executable_path='C:/Users/user/.wdm/drivers/chromedriver/win32/87.0.4280.88/chromedriver')

url = "https://www.google.com"

driver.get(url)
#-----------------------------구글켜기------------------------------------------
google = driver.find_element_by_name('q')# 구글검색창

google.send_keys('ssg')#구글검색창 입력
google.submit()#enter
#-----------------------------SSG------------------------------------------
driver.find_element_by_xpath('//*[@id="tads"]/div/div/div/div[1]/a/div[1]').click()#링크접속

ssg2= driver.find_element_by_class_name('inp_txt')#검색창
ssg2.click()
ssg2.send_keys('볶음밥')#입력

ssg3 = driver.find_element_by_id('ssg-query-btn')#enter
ssg3.click()
#----------------------------곱창중 첫번째 -------------------------------------------
time.sleep(1)

driver.find_element_by_xpath('//*[@id="item_unit_1000057096559"]/div[1]/div[2]/a').send_keys(Keys.ENTER)
#이동

#-----------------------------리뷰-------------------------------------------
review = driver.find_elements_by_css_selector('#cdtl_cmt_tbody>tr')#리뷰 파싱 
review_data_list=[]

#——————————————페이지1번부터10번까지————————————————————
for i in tqdm(range(1,10+1)):
    f = '//*[@id="comment_navi_area"]/a[{}]'.format(str(i))
    print(i,'=int',driver.find_element_by_xpath(f).text)
    time.sleep(1)
    f = '//*[@id="comment_navi_area"]/a[{}]'.format(str(i))
    print(driver.find_element_by_xpath(f).text)
    driver.find_element_by_xpath(f).send_keys(Keys.ENTER)
    time.sleep(3)
    
    rev = driver.find_elements_by_css_selector('#cdtl_cmt_tbody>tr') #처음 정의했던걸 다시 정의하고 시작하니 되는거같습니다!!
    
    for review_data in tqdm(rev):
        review_data_list.append(review_data.text.split('\n'))
        print('—————————2')
        time.sleep(3)
#————————————————페이지 n번에 있는 리뷰 10개 불러오기——————————————
for h in range(1,500):
    for i in tqdm(range(3,12+1)):
        time.sleep(1)
        f = '//*[@id="comment_navi_area"]/a[{}]'.format(str(i))
        print(driver.find_element_by_xpath(f).text)
        driver.find_element_by_xpath(f).send_keys(Keys.ENTER)
        time.sleep(3)

        rev = driver.find_elements_by_css_selector('#cdtl_cmt_tbody>tr') #처음 정의했던걸 다시 정의하고 시작하니 되는거같습니다!!

        for review_data in tqdm(rev):
            print(review_data.text.split('\n'))
            review_data_list.append(review_data.text.split('\n'))
            print('—————————2')
            time.sleep(3)
 
```
