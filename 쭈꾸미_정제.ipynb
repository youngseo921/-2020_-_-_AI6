```python
import pandas as pd 
import numpy as np 
jjuggu=pd.read_csv('ssg_jjuggu.csv',encoding ="utf-16")
a= []
for i in range(len(jjuggu['후기'])):
    a.append(jjuggu['후기'][i])
```

```python
# 토큰화
from konlpy.tag import Hannanum
h = []
hannanum = Hannanum()
for i in range(len(a)):
    h.append(hannanum.morphs(a[i]))
print(h) 
```

```python
# 태깅
hannanumTag= []
for i in range(len(h)):
    for token in h[i]:        
        hannanumTag += hannanum.pos(token) 
print(hannanumTag)
```

```python
#빈도단어
from collections import Counter
Counter(hannanumTag).most_common()
```

```python
#불용어 처리
stopPos = ['N','M']
stopWord = ['ㅠㅠㅠㅠ','다가','ㅎㅎ','지만','이랑','여기','ㅎㅎㅎ','그렇다치','었으','겠어요','ㄴ지','ㅂ니','ㅠㅠ','ㄴ데','어요','아요','ㄴ다','ㄴ까','었는데','었어요','었습니','ㅎㅎㅎㅎ','었네요','네요','이거','너무','자주','아아','근데','ㄹ께요']
쭈꾸미 = []
for tag in hannanumTag:
    if tag[1] in stopPos:
        if tag[0] not in stopWord:        
            if len(tag[0])>1:
                쭈꾸미.append(tag[0])  
print(쭈꾸미)
```

```python
word = str(쭈꾸미).replace(',','').replace('[','').replace(']','').replace("'",'')
#gensim 키워드 추출
from gensim.summarization import keywords
test2 = keywords(word,words = 40).split()
print(test2)
```
