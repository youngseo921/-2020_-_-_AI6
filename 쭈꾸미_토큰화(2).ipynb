```python
from konlpy.tag import Komoran

c= []
komoran = Komoran()

for i in range(len(a)):
      c.append(komoran.morphs(a[i]))
print(c)
```

```python
# 이중리스트
```

```python
 def flatten(list):
        result = []
        for item in c:
            if type(item) ==list:
                result+= flatten(item)
            else:
                result+=[item]
        return result
```

```python
flatten(c)
```

```python
list_of_list = c
result = []
for item in list_of_list:
    result.extend(item)
c = result
```

```python
c
```

```python
komoranTag = []

for token in c:
    komoranTag+= komoran.pos(token)
print(komoranTag)
```
