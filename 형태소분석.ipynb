# Okt
### 모든 리뷰를 Okt로 토큰화 한 경우 

```python
from konlpy.tag import Okt
```

```python
okt_tokens =[] 
okt= Okt()
for i in range(len(a)):
    okt_tokens += okt.morphs(a[i])
print(okt_tokens) 
```

```python
oktTag = []
for token in okt_tokens:
    oktTag += okt.pos(token)
print(oktTag)
```

### okt로 리뷰데이터에서 명사만 추출

```python
#okt로 리뷰데이터에서 명사만 추출
o_nouns = []
for i in range(len(a)):
    o_nouns += okt.nouns(a[i])
```

```python
o_nouns
```

```python
# 추출한 명사를 nltk 활용해서 text로 만들었을때 가장 많이 사용된 언어 150개 
# 리스트로 만들기 전 
```

```python
import nltk
okt_nouns_darg= nltk.Text(o_nouns)
```

```python
okt_nouns_darg.vocab().most_common(20)
```

```python
my_stopword = [] 
okt_list = []
for word in o_nouns_darg:
    if word not in my_stopword:
        okt_list.append(word)
```

```python
okyt_list
```

```python
#닭가슴살 명사 추출 텍스트 내 연관검색어 
o_nouns_darg.collocations()
```
